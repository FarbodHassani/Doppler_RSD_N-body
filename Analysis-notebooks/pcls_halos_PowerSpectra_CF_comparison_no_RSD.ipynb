# Powerspectra comparison

```python
plt.rc('text', usetex=True)
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : text_size}
matplotlib.rc('font', **font)
plt.figure(figsize=(10,8))

for i in [0]:
    plt.loglog(gev_lcdm[i][0,:],gev_lcdm[i][1,:],".",color =ColorsI[i], lw=3,label="$\Lambda$CDM, CDM, z="+redshifts[i])
# plt.loglog(P1k_kess_csem7_z1[0,:],P1k_kess_csem7_z1[1,:],".",lw=3,label="$k$-essence, $c_s^2=10^{-7}$")
    plt.loglog(gev_lcdm_h[i][0,:],gev_lcdm_h[i][1,:],".",color =ColorsI[i+1], lw=3,label="$\Lambda$CDM, halo, z="+redshifts[i])
    plt.loglog(gev_cs2_1[i][0,:],gev_cs2_1[i][1,:],".",color =ColorsI[i+2], lw=3,label="$c_s^2 = 1$, z="+redshifts[i])
    plt.loglog(gev_cs2_1_h[i][0,:],gev_cs2_1_h[i][1,:],".",color =ColorsI[i+3], lw=3,label="$c_s^2 = 1$, z="+redshifts[i])


    
    # plt.ylim(1.e1,1.e7)
    plt.ylabel('$P(k)$',fontsize=24)
    plt.xlabel('$k[h/Mpc]$',fontsize=24)
    # plt.yscale('log')
    plt.xticks(size = 25);
    plt.yticks(size = 25);
#     plt.ylim(1.e4,3.e5)
    plt.xlim(1.e-3,2.e0)
plt.legend(fontsize=20,bbox_to_anchor=(0.6,0.45))
```
