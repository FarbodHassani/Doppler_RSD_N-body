## Nested dictionary

```python
def nested_dict(n, type):
    if n == 1:
        return defaultdict(type)
    else:
        return defaultdict(lambda: nested_dict(n-1, type))
    
data = nested_dict(5, list)
pow_gev = nested_dict(2, list)
class_data = nested_dict(3, list)
rsd_coeff = nested_dict(2, list)
rsd_quad = nested_dict(2, list)
rsd_hexa = nested_dict(2, list)

rsd_coeff_halo = nested_dict(2, list)
rsd_quad_halo = nested_dict(2, list)
rsd_hexa_halo = nested_dict(2, list)

num_halos = nested_dict(2, list)

bias_h = nested_dict(2, list)
```
